```python
# =========================
# 1. 安裝套件
# =========================
!pip install -q gspread google-genai gradio pandas

# =========================
# 2. 匯入套件
# =========================
import gspread
import gradio as gr
import pandas as pd

from datetime import datetime
from google.colab import auth, userdata
from google.auth import default
from google import genai

# =========================
# 3. 基本設定
# =========================
SHEET_URL = "https://docs.google.com/spreadsheets/d/10ePR7tLFPu9VrIqIcg5Cvk945KCvsY1cBIWvvGfveLQ/edit?gid=1581224256#gid=1581224256"
WORKSHEET_NAME = "工作表2"
REQUIRED_COLUMNS = ["日期", "科目", "作業成績"]
MODEL_ID = "gemini-2.5-flash"

# 全域變數
_gc = None
_ws = None
_client = None


# =========================
# 4. 初始化 Gemini
# =========================
def init_gemini():
    global _client

    if _client is None:
        api_key = userdata.get("gemini")
        if not api_key:
            raise ValueError("找不到 Colab Secrets 裡的 gemini 金鑰，請先設定。")
        _client = genai.Client(api_key=api_key)

    return _client


# =========================
# 5. 初始化 Google Sheet
# =========================
def setup_gspread():
    global _gc, _ws

    if _ws is not None:
        return _ws

    auth.authenticate_user()
    creds, _ = default()
    _gc = gspread.authorize(creds)

    sh = _gc.open_by_url(SHEET_URL)
    _ws = sh.worksheet(WORKSHEET_NAME)

    ensure_headers()
    return _ws


def ensure_headers():
    global _ws

    if _ws is None:
        raise ValueError("工作表尚未初始化。")

    first_row = _ws.row_values(1)

    # 若第一列是空的，就補上表頭
    if not first_row:
        _ws.append_row(REQUIRED_COLUMNS)
        return

    # 若表頭不一致，就直接覆蓋 A1:C1
    if first_row[:len(REQUIRED_COLUMNS)] != REQUIRED_COLUMNS:
        _ws.update("A1:C1", [REQUIRED_COLUMNS])


# =========================
# 6. 清理輸入資料
# =========================
def clean_grade_data(grade_data):
    """
    將 Gradio Dataframe 傳入資料清理成：
    [["國文", 85], ["英文", 92]]
    """
    cleaned = []

    if grade_data is None:
        return cleaned

    for row in grade_data:
        if row is None or len(row) < 2:
            continue

        subject = "" if row[0] is None else str(row[0]).strip()
        grade_text = "" if row[1] is None else str(row[1]).strip()

        # 完全空白列直接略過
        if subject == "" and grade_text == "":
            continue

        if subject == "":
            raise ValueError("有一筆資料缺少科目。")

        if grade_text == "":
            raise ValueError(f"科目「{subject}」的成績不能空白。")

        try:
            grade = int(float(grade_text))
        except Exception:
            raise ValueError(f"科目「{subject}」的成績「{grade_text}」不是有效數字。")

        cleaned.append([subject, grade])

    return cleaned


# =========================
# 7. 建立 AI 提示詞
# =========================
def build_prompt(grade_rows):
    """
    grade_rows 格式：
    [["2026-04-06", "國文", 85], ["2026-04-06", "英文", 92]]
    """
    lines = [
        "以下是學生的成績資料，請幫我整理：",
        "1. 整體表現摘要",
        "2. 各科表現觀察",
        "3. 可能的常見迷思或學習提醒",
        "請用簡單、清楚、不評分的方式說明。",
        ""
    ]

    for date_str, subject, grade in grade_rows:
        lines.append(f"日期：{date_str}，科目：{subject}，成績：{grade}")

    return "\n".join(lines)


# =========================
# 8. 產生 AI 摘要
# =========================
def get_ai_summary(grade_rows):
    client = init_gemini()
    prompt_text = build_prompt(grade_rows)

    response = client.models.generate_content(
        model=MODEL_ID,
        contents=prompt_text
    )

    summary = getattr(response, "text", None)
    if not summary:
        return "AI 沒有成功產生摘要。"

    return summary.strip()


# =========================
# 9. 將 AI 摘要寫入 Google Sheet
# =========================
def write_summary_to_sheet(summary):
    global _ws

    if _ws is None:
        raise ValueError("Google Sheet 尚未連線。")

    today = datetime.now().strftime("%Y-%m-%d")

    summary_lines = [line.strip() for line in summary.split("\n") if line.strip()]
    if not summary_lines:
        summary_lines = ["AI 未產生可寫入的摘要內容。"]

    start_row = len(_ws.get_all_values()) + 1

    rows_to_write = []
    for i, line in enumerate(summary_lines):
        if i == 0:
            rows_to_write.append([today, "AI 摘要", line])
        else:
            rows_to_write.append(["", "", line])

    end_row = start_row + len(rows_to_write) - 1
    _ws.update(f"A{start_row}:C{end_row}", rows_to_write)


# =========================
# 10. 主處理函式
# =========================
def process_grades_and_summary(grade_data):
    global _ws

    # 確保 Google Sheet 連線成功
    try:
        setup_gspread()
    except Exception as e:
        return f"Google Sheet 連線失敗：{e}", ""

    # 清理與檢查輸入
    try:
        cleaned_data = clean_grade_data(grade_data)
    except Exception as e:
        return f"資料格式錯誤：{e}", ""

    if not cleaned_data:
        return "沒有輸入任何有效成績，請至少輸入一筆科目和成績。", ""

    # 整理成要寫入 Google Sheet 的格式
    today = datetime.now().strftime("%Y-%m-%d")
    grade_rows_for_sheet = [[today, subject, grade] for subject, grade in cleaned_data]

    # 寫入成績
    try:
        _ws.append_rows(grade_rows_for_sheet)
        sheet_message = "成績已成功寫入 Google Sheet。"
    except Exception as e:
        return f"寫入 Google Sheet 失敗：{e}", ""

    # 生成 AI 摘要
    try:
        summary = get_ai_summary(grade_rows_for_sheet)
    except Exception as e:
        return sheet_message + f"\n但 AI 摘要生成失敗：{e}", ""

    # 寫入 AI 摘要到 Google Sheet
    try:
        write_summary_to_sheet(summary)
        sheet_message += "\nAI 摘要已成功寫入 Google Sheet。"
    except Exception as e:
        sheet_message += f"\nAI 摘要已生成，但寫入 Google Sheet 失敗：{e}"

    return sheet_message, summary


# =========================
# 11. 啟動前測試初始化
# =========================
try:
    setup_gspread()
    init_gemini()
    print("初始化完成：Google Sheet 與 Gemini 已準備就緒。")
except Exception as e:
    print(f"初始化失敗：{e}")


# =========================
# 12. 建立 Gradio 介面
# =========================
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("# 成績輸入與 AI 摘要工具")
    gr.Markdown("請在下方輸入學生的科目與成績，送出後會寫入 Google Sheet，並產生 AI 摘要。")

    with gr.Row():
        with gr.Column():
            grade_input = gr.Dataframe(
                headers=["科目", "成績"],
                value=[["", ""]],
                type="array",
                row_count=3,
                col_count=(2, "fixed"),
                label="輸入科目與成績"
            )
            submit_button = gr.Button("送出", variant="primary")

        with gr.Column():
            sheet_output = gr.Textbox(label="Google Sheet 處理狀態", lines=4)
            summary_output = gr.Textbox(label="AI 摘要", lines=15)

    submit_button.click(
        fn=process_grades_and_summary,
        inputs=grade_input,
        outputs=[sheet_output, summary_output]
    )

demo.launch(debug=True)
```
